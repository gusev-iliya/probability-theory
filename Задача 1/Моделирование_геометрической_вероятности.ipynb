#Задача
На квадратном листе бумаги со стороной
10 см нарисован равнобедренный прямоугольный треугольник так, что его катеты лежат на двух смежных сторонах квадрата и имеют длину 6 см. Точку на листе выбирают наугад равномерно по всей площади квадрата.
Какова вероятность того, что выбранная точка окажется внутри этого треугольника?



#Теоретический расчет
Площадь листа бумаги
$$ Sкв = 0.1\cdot 0.1 = 0.01$$
Площадь треугольника
$$ Sтр = \frac{0.06 \cdot 0.06}{2} = 0.0018$$
Вероятность, что случайно выбранная точка окажется внутри треугольника
$$ P(A) = \frac{Sтр}{Sкв} = \frac{0.0018}{0.01} = 0.18$$

#Графическая иллюстрация


```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Polygon
import numpy as np
```

```python
def Probs(n):
    a = 0.1
    fig, ax = plt.subplots(figsize=(5, 5))
    ax = plt.gca()
    ax.add_patch(Rectangle((0, 0), a, a, facecolor='green'))
    ax.add_patch(Polygon([[0, 0], [0, 0.06], [0.06, 0]], facecolor='blue'))
    x = np.random.uniform(0, a, n)
    y = np.random.uniform(0, a, n)
    plt.scatter(x, y, marker='.', c='black')

    arr = []
    ins=0
    for i in range(len(x)):
        if (y[i] <= 0.06 - x[i] ):
            plt.scatter(x[i], y[i], marker='.', c='red')
            ins+=1
    prob = ins / n
    plt.show()
    print('В треугольник попало %s точек. Вероятность попадания = %s' %(ins,prob))
```

```python
Probs(3000)
```
