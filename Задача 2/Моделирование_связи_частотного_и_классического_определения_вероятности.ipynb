#Задача
Из стандартной игральной колоды (36 карт) по очереди достают по одной карте. Какова вероятность достать 4 туза подряд?
#Классическая вероятность
Событие А - из колоды достали 4 туза подряд.
Вероятность Pi достать из колоды туз на каждом последующем шаге:

$P1 = \frac{C_{4}^{1}}{C_{36}^{1}} = \frac{4}{36}= \frac{1}{9}$

$P2 = \frac{C_{3}^{1}}{C_{35}^{1}} = \frac{3}{35}$

$P3 = \frac{C_{2}^{1}}{C_{34}^{1}} = \frac{2}{34}= \frac{1}{17}$

$P4 = \frac{C_{1}^{1}}{C_{33}^{1}} = \frac{1}{33}$

$P(A) = P1 \cdot P2 \cdot P3 \cdot P4 =  \frac{1}{9} \cdot \frac{3}{35} \cdot \frac{1}{17} \cdot \frac{1}{33} = 0.00001697$

#Частотная вероятность

```python
import random
import plotly.graph_objs as go
import numpy as np
deck = ['6']*4 + ['7']*4+['8']*4+['9']*4+['10']*4+['J']*4+['Q']*4+['K']*4+['A']*4

n = 10000000
N_A = 0
probs = [0]
for i in range(n//500):
  for j in range(500):
    if random.sample(deck,4) == ['A','A','A','A']:
      N_A += 1
  probs.append(N_A/(i*500 + 500))
```

```python
print('Количество раз когда были вытянуты 4 туза подряд при %s испытаний:%s'% (n,N_A))
print("Частота при n = %s: %s" %(n, probs[-1]))
N_values = np.array(range(0, n, 500))
P = [0.00001697] * len(N_values)

fig = go.Figure(layout=dict(width=600, height=300))
fig.update_layout(margin=dict(l=0, t=0, b=0))

fig.add_trace(go.Scatter(x=N_values, y=probs, name='Частота'))
fig.add_trace(go.Scatter(x=N_values, y=P, name='Вероятность', line_dash='dash'))

fig.update_xaxes(title_text='N')
fig.update_yaxes(title_text='P(A)')
fig.show()
```
